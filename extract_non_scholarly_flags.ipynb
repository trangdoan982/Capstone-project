```python
import pandas as pd

# Read the fake new lists and clean up the names
red_news = []
data = pd.read_csv("fake_news_list.csv", header=None)
for ele in data.iloc[:, 0]:
    a = ele.split("(")
    red_news.append(a[0].strip())

red_news
        
```

Scrape from [Perennial Sources list](https://en.wikipedia.org/wiki/Wikipedia:Reliable_sources/Perennial_sources)

```python
mapping ={
    "s-gr": "green", #generally reliable
    "s-nc": "yellow", #no consensus
    "s-gu": "red", #generally unreliable
    "s-d": "red", #deprecated
    "s-b": "red" #blacklisted
}
```

```python
import requests
from bs4 import BeautifulSoup

non_scholarly_flags = {"red": [], "green":[], "yellow": []}

url = "https://en.wikipedia.org/wiki/Wikipedia:Reliable_sources/Perennial_sources"
res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")
table = soup.find("table", {"class": "wikitable"})
body = table.find("tbody")
all_rows = body.find_all("tr")
for row in all_rows:
    if row.has_attr("id"):
        name = row["id"].replace("_", " ")
        name = name.split("(")[0]
        tag = row["class"][0]
        flag = mapping[tag]
        non_scholarly_flags[flag].append(name)

non_scholarly_flags
```

```python
#add the fake news list to the dictionary
non_scholarly_flags["red"] += red_news

non_scholarly_flags
```

```python
# Export to a JSON file
import json

with open("non_scholarly_flags.json", "w") as outfile:
    json.dump(non_scholarly_flags, outfile)
```
